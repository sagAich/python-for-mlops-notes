```python
# we are creating a function that checks if the input is negative or not. If it is negative, it raises a ValueError with a custom message. 
# If it is not negative, it returns the input value.
def check_input(a):
    if a < 0:
        raise ValueError("Negative value is not allowed")
    return a
```

```python
check_input(-5)
```

```python
# now instead of getting long error message, we can use try except block to catch the error and print only the custom message.
try: 
    check_input(-5)
except ValueError as e:
    print(e)
```

```python
check_input(10)
```

```python
# now instead of valueerror or any other built in exception, let's define our own custom exception.
def check_input(a):
    if a < 0:
        raise xyz("Negative value is not allowed")
```

```python
try:
    check_input(-5)
except ValueError as e:
    print(f"An error occurred: {e}")
```

```python
# As you can see, the last cell raises a NameError because 'xyz' is not defined. Now following is the way to define a custom exception:
class xyz(Exception): # here Exception is the super class.
    pass
```

```python
def register(age):
    if age<0:
        raise xyz("Age cannot be negative")
```

```python
try:
    register(-5)
except xyz as e:
    print(e)
```

```python
class Tooshortpassword(Exception):
    def __init__(self, password_length, message="Password is too short"):
        # we are defining a constructor so that class can take two arguments, 
        # password_length and message as input.
        self.password_length = password_length
        self.message = message
        super().__init__(self.message)
        # It calls the parent Exception class's constructor/dunder method (__init__), 
        # passing it your custom message so Python knows exactly what to display when the exception is raised.
```

```python
def check_password(password):
    if len(password)<8:
        raise Tooshortpassword(len(password))
    print("Password is valid")
```

```python
try:
    check_password("abc")
except Tooshortpassword as e:
    print(f"Password length is {e.password_length}. {e.message}")
```
